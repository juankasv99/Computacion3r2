# AVALUACIÓ - Problema avaluació de models

## Cas binari

### Corbes ROC i PR

Primerament fem els imports necessaris:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,  precision_score, recall_score, roc_curve,roc_auc_score, auc
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, precision_recall_curve
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

Obrim el dataset i preparem un objecte StandardScaler per estandaritzar el dataset posteriorment:

```python
datasetPath = './Data/fashion-mnist_train_small_binary.csv'
dataFrame = pd.read_csv(datasetPath)
stdScaler = StandardScaler()
```

Separem els valors en X (valor dels pixels) i Y (lables):

```python
x = dataFrame.drop('label', axis = 1)
y = dataFrame[['label']]
```

Visualitzem la distribució dels labels:

```python
y.hist()
plt.show();
```

En aquest cas, els labels són binaris i podemo obvservar com es troben clarament desbalançejats, hi ha molts 1s i pocs 0s.

Estandaritzem els valors d'X:

```python
stdScaler.fit(x)
x = pd.DataFrame(stdScaler.transform(x))
x.columns = dataFrame.columns[1:]
```

Separem les dades en train i test:

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3, random_state=20)
print(f'Entrades de train: {xTrain.shape[0]}')
print(f'Entrades de test: {xTest.shape[0]}')
```

Per a cada valor de K del 1 al 15 de dos en dos visualitzem tant la corba ROC com la PR:

```python
for K in range(1, 16, 2):
    model = KNeighborsClassifier(n_neighbors=K)
    model.fit(xTrain, yTrain.values.ravel())
    predictions = model.predict(xTest)
    falsePositiveRate, truePositiveRate, _ = roc_curve(predictions, yTest)
    precision, recall, _ = precision_recall_curve(predictions, yTest)
    
    plt.figure(1)
    plt.plot(falsePositiveRate, truePositiveRate, marker='.', label=f'K={K}')
    
    plt.figure(2)
    plt.plot(recall, precision, marker='.', label=f'K={K}')

plt.figure(1)
plt.plot([0,1], [0,1], linestyle='--', color='black')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

plt.figure(2)
noSkill = yTest[yTest['label']==1].shape[0] / yTest.shape[0]
plt.plot([0, 1], [noSkill, noSkill], linestyle='--', color='black')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()

plt.show()
```

Sembla evident que en el nostre cas el millor valor per a K seria 7 ja que:  
1- Aconsegueix el TPR més alt amb el FPR més baix.  
2- Aconsegueix la precision més alta per al recall més alt.

### K-Fold Cross-validation

Utilitzem k-fold com a tècnica de cross-validation per comprobar que els resultats anteriors no han sigut solament sort:

```python
scores = []
for K in range(1, 16, 2):
    model = KNeighborsClassifier(n_neighbors=K)
    scores.append(cross_val_score(model, xTrain, yTrain.values.ravel(), cv=5, scoring='accuracy').mean())
plt.plot([K for K in range (1, 16, 2)], scores)
plt.xlabel('K value in KNN')
plt.ylabel('Cross-Validated accuracy')
plt.show()
```

Fem k-fold amb k=5 ja que en aquest cas tenim 26 entrades a cada partició, que equivalen al 20% de les 180 entrades que tenim a test. De tal manera que tenim un nomre ni massa gran ni massa petit.

Sembla ser que podem confirmar la nostra hipòtesis de que el K=7 al KNN és un bon valor.

## Cas multicategoria

### Corbes ROC

Obrim el dataset i preparem un objecte StandardScaler per estandaritzar el dataset posteriorment:

```python
datasetPath = './Data/fashion-mnist_train_small.csv'
dataFrame = pd.read_csv(datasetPath)
stdScaler = StandardScaler()
```

Separem els valors en X (valor dels pixels) i Y (lables):

```python
x = dataFrame.drop('label', axis = 1)
y = dataFrame[['label']]

print(y['label'].unique())
```

En aquest cas de clasificació multicategoría veiem que tenim 10 clases (del 0 al 9) diferents.

Visualitzem la distribució dels labels:

```python
y.hist()
plt.show();
```

Sembla que en aquest cas están bastant ben distribuits els valors d'Y, lo qual és positiu.

Estandaritzem els valors d'X:

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3, random_state=20)
print(f'Entrades de train: {xTrain.shape[0]}')
print(f'Entrades de test: {xTest.shape[0]}')
```

Per a cada valor de K del 1 al 15 de dos en dos visualitzem tant la corba ROC com la PR:

```python
for K in range(1, 16, 2):
    model = KNeighborsClassifier(n_neighbors=K)
    model.fit(xTrain, yTrain.values.ravel())
    predictions = model.predict(xTest)

    tpr = []
    fpr = []
    for classIdx in range(0,10):
        class_predictions = np.array(list(map(lambda x: 1 if x==classIdx else 0, predictions)))
        class_yTest = yTest.applymap(lambda x: 1 if x==classIdx else 0)
        
        falsePositiveRate, truePositiveRate, _ = roc_curve(class_predictions, class_yTest)
        fpr.append(falsePositiveRate)
        tpr.append(truePositiveRate)
        
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(0, 10)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(0, 10):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= 10
    
    plt.plot(all_fpr, mean_tpr, marker='.', label=f'K={K}')

plt.plot([0,1], [0,1], linestyle='--', color='black')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

plt.show()
```

De nou, sembla ser que el valor K=7 és el més adequat.

### K-Fold Cross-validation

```python
scores = []
for K in range(1, 16, 2):
    model = KNeighborsClassifier(n_neighbors=K)
    scores.append(cross_val_score(model, xTrain, yTrain.values.ravel(), cv=3, scoring='accuracy').mean())
plt.plot([K for K in range (1, 16, 2)], scores)
plt.xlabel('K value in KNN')
plt.ylabel('Cross-Validated accuracy')
plt.show()
```

En aquest cas multicategoría, hem pensat que sería millor utilitzar particions més grans. Per tant, optem per una k=3 per a que tinguem una suficient representació de cada categoría a cada una de les particions.  
En aquest cas també sembla ser que escollir K=7 a KNN és una bona decisió.
